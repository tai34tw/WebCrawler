#### 處理df['詞名']頭尾空白
- 宋詞目錄

```python
# 去宋詞目錄之詞名的頭尾之空白
df = content_csv

## 處理前
print('處理前:*' + df['詞名'][0] + '*') # 用*確認空白格數
print('處理前:*' + df['詞名'][5] + '*') 
print("")

# 處理中
df = df.applymap(lambda x: str(x)) # 所有cell data str化
df = df.applymap(lambda x: x.lstrip()) # 去頭空白
print("去頭空白中...")
df = df.applymap(lambda x: x.rstrip()) # 去尾空白
print("去尾空白中...")
df = df.applymap(lambda x: x.replace("  "," ")) # 以單空白代替雙空白
print("以單空白代替雙空白中...")
                 
# 處理後
print("")
print('處理後:*' + df['詞名'][0] + '*') # 用*確認空白格數
print('處理後:*' + df['詞名'][5] + '*')

content_csv = df # 還回去
```

- 宋詞內文

```python
# 去宋詞內文之詞名的頭尾之空白
df = content_csv

## 處理前
print('處理前:*' + df['詞名'][0] + '*') # 用*確認空白格數
print('處理前:*' + df['詞名'][5] + '*') 
print("")

# 處理中
#df = df.applymap(lambda x: str(x)) # 所有cell data str化
df = df.applymap(lambda x: x.lstrip()) # 去頭空白
print("去頭空白中...")
df = df.applymap(lambda x: x.rstrip()) # 去尾空白
print("去尾空白中...")
df = df.applymap(lambda x: x.replace("  "," ")) # 以單空白代替雙空白
print("以單空白代替雙空白中...")
                 
# 處理後
print("")
print('處理後:*' + df['詞名'][0] + '*') # 用*確認空白格數
print('處理後:*' + df['詞名'][5] + '*')

content_csv = df # 還回去
```

### 合併

```python
SongShi = pd.merge(index_csv, content_csv, how='outer', on=['網址'], indicator = True)
SongShi.tail(1)
```

```python
# 檢視有空值的欄位
SongShi.isnull().sum()
```

```python
# 檢視內文未合併完全的資料
SongShi[SongShi['_merge'] != 'both']
```

```python
# 檢視SongShi['作者_x']為空值處
SongShi[SongShi['作者_x'].isnull()]
```

```python
SongShi.iloc[131111:131112]
```

```python
# 補'作者_x'為作者_y
SongShi['作者_x'] = SongShi['作者_x'].fillna(0) # 先補零

for i in range(len(SongShi)):
    if SongShi['作者_x'][i]== 0:
        SongShi['作者_x'][i] = SongShi['作者_y'][i]
        
SongShi.iloc[131111:131113]
```

```python
# 去除不必要的columns
SongShi = SongShi.drop(columns=['id_x', 'id_y', '詞名_y', '作者_y'])

# 修改column名
SongShi = SongShi.rename(columns={'作者_x':'作者', '詞名_x':'詞名'})

SongShi.tail(2)
```

```python
# 修飾data
SongShi['id']=0 # 創建id欄位

for i in range (1, len(SongShi['id'])+1): # 修改id欄位，由小到大排序
    SongShi['id'][i] = i
    
# 調整欄位順序
SongShi = SongShi[['id', '詩名','卷號', '詩卷號', '作者', '詞名', '押韻', '體詩', '內容', 
                   '作者共做了幾首詩', '作者生平', '網址', '_merge']]

SongShi.head(2)
```

```python
SongShi.tail(2)
```

```python
# 將both, right_only 及 left_only隔離出來
both = SongShi[SongShi['_merge'] == 'both']
right = SongShi[SongShi['_merge'] == 'right_only']
left = SongShi[SongShi['_merge'] == 'left_only']

# 檢視各資料筆數
print("both:", len(both), ", right:", len(right),", left:", len(left))
print("total:", len(SongShi))
```
